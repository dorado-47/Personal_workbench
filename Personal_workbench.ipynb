```python
import mysql.connector
import tkinter as tk
from tkinter import messagebox

# Function to fetch databases
def show_databases():
    try:
        mydb = mysql.connector.connect(
            host="localhost",
            user="root",
            password="sd047"
        )
        cur = mydb.cursor()
        cur.execute("SHOW DATABASES")

        db_listbox.delete(0, tk.END)  # Clear previous list
        for db in cur:
            db_listbox.insert(tk.END, db[0])  # Insert databases into listbox

        cur.close()
        mydb.close()
    except mysql.connector.Error as e:
        messagebox.showerror("Error", f"Database Connection Failed!\n{e}")

# Function to fetch tables of selected database
def show_tables():
    selected_db = db_listbox.get(tk.ACTIVE)  # Get selected database
    if not selected_db:
        messagebox.showwarning("Warning", "Please select a database first!")
        return
    
    try:
        mydb = mysql.connector.connect(
            host="localhost",
            user="root",
            password="2005",
            database=selected_db
        )
        cur = mydb.cursor()
        cur.execute("SHOW TABLES")

        table_listbox.delete(0, tk.END)  # Clear previous list
        for table in cur:
            table_listbox.insert(tk.END, table[0])  # Insert tables into listbox

        cur.close()
        mydb.close()
    except mysql.connector.Error as e:
        messagebox.showerror("Error", f"Failed to retrieve tables!\n{e}")

# GUI setup
root = tk.Tk()
root.title("MySQL Database & Table Viewer")
root.geometry("500x500")
root.configure(bg="#f4f4f4")

# Title Label
title_label = tk.Label(root, text="MySQL Database Viewer", font=("Arial", 14, "bold"), bg="#f4f4f4")
title_label.pack(pady=10)

# Databases Section
db_label = tk.Label(root, text="Databases:", font=("Arial", 12, "bold"), bg="#f4f4f4")
db_label.pack()
db_listbox = tk.Listbox(root, width=50, height=8, font=("Arial", 12))
db_listbox.pack(pady=5)

show_db_btn = tk.Button(root, text="Show Databases", font=("Arial", 12, "bold"), bg="#4CAF50", fg="white",
                        command=show_databases)
show_db_btn.pack(pady=5)

# Tables Section
table_label = tk.Label(root, text="Tables:", font=("Arial", 12, "bold"), bg="#f4f4f4")
table_label.pack()
table_listbox = tk.Listbox(root, width=50, height=8, font=("Arial", 12))
table_listbox.pack(pady=5)

show_table_btn = tk.Button(root, text="Show Tables", font=("Arial", 12, "bold"), bg="#2196F3", fg="white",
                           command=show_tables)
show_table_btn.pack(pady=5)

root.mainloop()

```
